**Task 14:**

Store the user input for the City in the variable named as "Input_City3". Text to be shown to the user "Enter the City Name : "

Store the user input for the Payment in the variable named as "Input_Payment2". Text to be shown to the user "Enter the Payment : " Filter the sales dataframe for City to be taken from user input and Payment to be taken from user input. Store the resultant data in "Sales4" dataframe.

Create a pivot table using the Sales4 dataframe named pivot_2 to show the Productline-wise total GrossIncome and total Quantity for each of the Order. Use index as "Productline", columns as "Customer_type" and values as "GrossIncome" and "Quantity". Print this pivot table. This will show the data for the City and Payment selected by the user.

At the end the user input should be-

City (Input_City3) - Enter the City name : Yangon

Payment (Input_Payment2) - Enter the Payment : Cash

```python
Input_City3=input("Enter the City Name : ")
Input_Payment2=input("Enter the Payment : ")
Sales4=salesdata.query('City==@Input_City3 and Payment==@Input_Payment2')
pivot_2=pd.pivot_table(Sales4,index=["Productline"], columns=["Customer_type"],values=["GrossIncome","Quantity"],aggfunc={"GrossIncome":["sum"],"Quantity":["sum"]})
pivot_2
```
