***`Task 1: Import Pandas Library and create a dataframe named “tyresales” from a given URL and then print it. `***
URL - https://raw.githubusercontent.com/Invact-Abhay/DO9/main/tyresale.csv

```python
import pandas as pd
tyresales=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DO9/main/tyresale.csv")
tyresales
```

***`Task 2: Clean the customer name by using strip and title function and removing '@' and any space between the name and give name to this column - ‘CustomerNameClean’ then retrieve ‘CustomerNameClean’ in Alphabetical order. ‘CustomerNameClean’ should look like this - AlphaMotorsPvtLtd.`***

```python
tyresales["CustomerNameClean"]=tyresales["CustomerName"].str.strip().str.title().str.replace('@','').str.replace(' ','')
tyresales.sort_values('CustomerNameClean',ascending=True)["CustomerNameClean"]
tyresales
```

***`Task 3: Calculate the Sales for every customer by calculating Sales column.`***

```python
sales_for_customers=tyresales.groupby('CustomerNameClean')["SalesCommissionPaid"].sum()
sales_for_customers
```

***`Task 4: Find the total Sales.`***

```python
total_sales=tyresales["SalesCommissionPaid"].sum()
total_sales
```

***`Task 5: Take the user Input for CustomerNameClean and retrieve the data for 'Alpha Motors Pvt Ltd'.`***

```python
import pandas as pd
import re
# Define a function to split at title-case words and rejoin them without commas
def split_title_words(name):
    # Split based on title-case words
    words = re.findall(r'[A-Z][a-z]*', name)
    # Rejoin the words without commas, using space as a separator
    return ' '.join(words)

# Apply the function to the CustomerName column
tyresales['CustomerNameSplit'] = tyresales['CustomerNameClean'].apply(split_title_words)

#print(tyresales)
#tyresales[['CustomerNameSplit','CustomerNameClean']]
# Take user input for CustomerNameClean
customer_input = input("Enter the Customer Name (e.g., AlphaMotorsPvtLtd): ")

# Retrieve data for the inputted customer name
retrieve_data = tyresales.query('CustomerNameClean== @customer_input')

# Check if the customer exists in the data and print the result
if not retrieve_data.empty:
    print(retrieve_data)
else:
    print(f"No data found for customer: {customer_input}")
```

***`Task 6: Find the average SalesCommission% by calculating the SalesCommission% column.`***

SalesCommission% = SalesCommissionPaid / Sales * 100

```python
tyresales['Sales']=tyresales['PerUnitPrice']*tyresales['UnitsSold']
tyresales['SalesCommission%']=tyresales['SalesCommissionPaid']/tyresales['Sales']*100
tyresales['SalesCommission%'].mean()
```

**`Task 7: Take the user Input for Zone and Find the average SalesCommission% of Central Zone.`**

```python
# Take user input for Zone
user_zone = input("Enter the Zone (e.g., Central, West, East, South, North): ")

# Filter the DataFrame based on the user input
df_filtered = tyresales.query('Zone == @user_zone')

# Calculate the average SalesCommission% for the selected Zone
if not df_filtered.empty:
    avg_commission = df_filtered['SalesCommission%'].mean()
    print(f"Average SalesCommission% for {user_zone} Zone: {avg_commission}%")
else:
    print(f"No data available for the {user_zone} Zone.")
```
