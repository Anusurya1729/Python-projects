**Task 1**


Import Pandas  and matplotlib Library.
Create a dataframe named "ytdata from a given URL and then print it.

URL - https://raw.githubusercontent.com/Invact-Abhay/DOE/main/YTDATADOE11.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
ytdata=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/main/YTDATADOE11.csv")
ytdata
```

**Task 2**

Convert impression, views, comment, and likes into lakh (in '00000 format) and round to 2 decimal places  and name these columns  -


*   impression(in lakhs)
*   views(in lakhs)
*   comment(in lakhs)
*   likes(in lakhs)


Convert duration_seconds to duration_minutes and round to 2 decimal places.

Display the 5 rows of DataFrame

```python
ytdata["impression(in lakhs)"]=(ytdata['impression']/100000).round(2)
ytdata["views(in lakhs)"]=(ytdata['views']/100000).round(2)
ytdata["comment(in lakhs)"]=(ytdata['comment']/100000).round(2)
ytdata["likes(in lakhs)"]=(ytdata['likes']/100000).round(2)
ytdata["duration_minutes"]=(ytdata['duration_seconds']/100000).round(2)
ytdata.head()
```

**Task 3:**

Extract the year and month from publishTime in separate column named "year" and "month" respectively.

Display the 5 rows of DataFrame

```python
ytdata['publishTime']=pd.to_datetime(ytdata['publishTime'])
ytdata['year']=ytdata['publishTime'].dt.year
ytdata['month']=ytdata['publishTime'].dt.month
ytdata.head()
```

**Task 4:**

Drop the following columns -
'impression', 'views', 'publishTime', 'comment', 'likes', 'duration_seconds'

Display the 5 rows of DataFrame

```python
ytdata.drop(columns=["duration_seconds","views","likes","comment","impression","publishTime"],inplace=True)
ytdata.head(5)
```

# Bar Chart

**Task 5:**


Create a pivot table from the ytdata DataFrame that counts the number of videos (title) for each year. Store this pivot table in ytDatapivotbar and reset its index. Then, print the ytDatapivotbar.

```python
ytDatapivotbar=pd.pivot_table(ytdata,index=["year"],values=["title"],aggfunc={"title":"count"})
ytDatapivotbar.reset_index(inplace=True)
ytDatapivotbar
```


**Task 6:**

Create a bar plot using the ytDatapivotbar DataFrame, with years on the x-axis and the count of videos on the y-axis. Then, display the plot.

```python
plt.bar(x=ytDatapivotbar["year"],height=ytDatapivotbar["title"])
plt.show()
```

**Task 7:**

Create a bar plot using the ytDatapivotbar DataFrame, with years on the x-axis and the count of videos on the y-axis.

Label the x-axis as "Year"

and the y-axis as "Count of videos".

Add the title "Yearly wise videos uploaded",

and rotate the x-axis labels by 45 degrees.

Then, display the plot.

```python
plt.bar(x=ytDatapivotbar["year"],height=ytDatapivotbar["title"])
plt.xlabel("year")
plt.ylabel("Count of videos")
plt.title("Yearly wise videos uploaded")
plt.xticks(ticks=ytDatapivotbar["year"],rotation=45)
plt.show()
```

# Scattere Plot

**Task 17:**

Create a scatter plot to visualize the relationship between views and impression for videos with publish_country as "US" in the ytdata DataFrame. Then, display the plot.

```python
US_viewCount=ytdata.query('publish_country=="US"')["views(in lakhs)"]
US_impressionCount=ytdata.query('publish_country=="US"')["impression(in lakhs)"]
plt.scatter(US_viewCount,US_impressionCount)
plt.show()
```

**Task 18:**

Create a scatter plot to visualize the relationship between views and impression for videos with publish_country as "US" and "IN" in the ytdata DataFrame. Then, display the plot.








```python
US_viewCount=ytdata.query('publish_country=="US"')["views(in lakhs)"]
US_impressionCount=ytdata.query('publish_country=="US"')["impression(in lakhs)"]
plt.scatter(US_viewCount,US_impressionCount)
IN_viewCount=ytdata.query('publish_country=="IN"')["views(in lakhs)"]
IN_impressionCount=ytdata.query('publish_country=="IN"')["impression(in lakhs)"]
plt.scatter(IN_viewCount,IN_impressionCount)
plt.show()
```

**Task 19:**


Add the label and titles in the scatter plot created above as per the details provided below:

For US (Publish Country) scatter plot -   label - 'US'

For IN  (Publish Country) scatter plot -   label - 'IN'

Label the x-axis as "View Count(in lakhs)"

and the y-axis as "Impression Count(in lakhs)"

Add the title "US vs IN".

and place the legend on upper left and legend title is 'Publish Country'

```python
US_viewCount=ytdata.query('publish_country=="US"')["views(in lakhs)"]
US_impressionCount=ytdata.query('publish_country=="US"')["impression(in lakhs)"]
plt.scatter(US_viewCount,US_impressionCount,label="US")
IN_viewCount=ytdata.query('publish_country=="IN"')["views(in lakhs)"]
IN_impressionCount=ytdata.query('publish_country=="IN"')["impression(in lakhs)"]
plt.scatter(IN_viewCount,IN_impressionCount,label="IN")
plt.xlabel("View Count(in lakhs)")
plt.ylabel("Impression Count(in lakhs)")
plt.title("US vs IN")
plt.legend(title="legend",loc="upper left")
plt.show()
```
