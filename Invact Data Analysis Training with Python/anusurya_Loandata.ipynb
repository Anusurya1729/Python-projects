**Task 1**

Import Pandas and matplotlib Library. Create a dataframe named "LoanData" from a given URL and then print it.

URL - https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/Assignment%20DOE11.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
LoanData=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/Assignment%20DOE11.csv")
LoanData
```

**Task 2:**

Extract the year and month from "issuedate" in separate column named "year" and "month" respectively.

Display the 5 rows of DataFrame


```python
LoanData["issuedate"]=pd.to_datetime(LoanData["issuedate"])
LoanData["month"]=LoanData["issuedate"].dt.month
LoanData["year"]=LoanData["issuedate"].dt.year
LoanData.head(5)
```

**Task 3:**

Drop the following columns - 'memberid', 'emptitle' and 'subgrade'

Display the 5 rows of DataFrame

```python
LoanData.drop(columns=['memberid','emptitle','subgrade'],inplace=True)
LoanData.head(5)
```

**Task 4:**

Create a pivot table from the "LoanData" DataFrame that counts the number of loans (id) for each purpose. Store this pivot table in LoanPivotBar and reset its index. Then, print the LoanPivotBar.

```python
LoanPivotBar=pd.pivot_table(LoanData,index=['purpose'],values=['id'],aggfunc={"id":"count"})
LoanPivotBar.reset_index(inplace=True)
LoanPivotBar
```

**Task 5:**

Create a bar plot using the LoanPivotBar DataFrame, with purpose on the x-axis and the count of loan id on the y-axis. Then, display the plot.

```python
plt.bar(x=LoanPivotBar['purpose'],height=LoanPivotBar['id'])
plt.show()
```

**Task 6:**

Create a bar plot using the LoanPivotBar DataFrame, with purpose on the x-axis and the count of loan id on the y-axis

Label the x-axis as "Loan Purpose"

and the y-axis as "Number of Loan".

Add the title "Purpose Wise Loan Count",

and rotate the x-axis labels by 90 degrees.

Then, display the plot.

```python
plt.bar(x=LoanPivotBar['purpose'],height=LoanPivotBar['id'])
plt.xlabel("Loan Purpose")
plt.ylabel("Number of Loan")
plt.title("Purpose Wise Loan Count")
plt.xticks(ticks=LoanPivotBar['purpose'],rotation=90)
plt.show()
```

# Pie Chart

**Task 7:**

Create a pivot table from the "LoanData" DataFrame for loanstatus wise loanamount.

Store this pivot table in loanPivotPie and reset its index.

Display the resulting DataFrame.

```python
loanPivotPie=pd.pivot_table(LoanData,index=['loanstatus'],values=['loanamount'],aggfunc={'loanamount':'sum'})
loanPivotPie.reset_index(inplace=True)
loanPivotPie
```

**Task 8:**


Create a pie chart using the loanPivotPie DataFrame for loanstatus wise loanamount.

label the slices with loanstatus,

and set the title to "Loan Status".

Then, display the chart.

```python
plt.pie(x=loanPivotPie['loanamount'],labels=loanPivotPie['loanstatus'])
plt.title("Loan Status")
plt.show()
```

**Task 9:**

Create a pie chart using the loanPivotPie DataFrame for loanstatus wise loanamount.


label the slices with loanstatus,

and display the percentage of each slice with no decimal places.

Set the title to "Share of Loan Status"

and then display the chart.

```python
plt.pie(x=loanPivotPie['loanamount'],labels=loanPivotPie['loanstatus'],autopct="%.0f%%")
plt.title("Share of Loan Status")
plt.show()
```

# Scatter Plot

**Task 14:**

Create a scatter plot to visualize the relationship between loanamount and annualincome of loans with homeownership as "OWN" in the LoanData DataFrame. Then, display the plot.

```python
Own_loanamount=LoanData.query('homeownership=="OWN"')['loanamount']
Own_annualincome=LoanData.query('homeownership=="OWN"')['annualincome']
plt.scatter(Own_loanamount,Own_annualincome)
plt.show()
```

**Task 16:**

Create a scatter plot to visualize the relationship between loanamount and annualincome with homeownership as "OWN and "MORTGAGE" in the LoanData DataFrame. Then, display the plot.

```python
Own_loanamount=LoanData.query('homeownership=="OWN"')['loanamount']
Own_annualincome=LoanData.query('homeownership=="OWN"')['annualincome']
plt.scatter(Own_loanamount,Own_annualincome)
MORTGAGE_loanamount=LoanData.query('homeownership=="MORTGAGE"')['loanamount']
MORTGAGE_annualincome=LoanData.query('homeownership=="MORTGAGE"')['annualincome']
plt.scatter(MORTGAGE_loanamount,MORTGAGE_annualincome)
plt.show()
```

**Task 17:**

Add the label and titles in the scatter plot created above as per the details provided below:

For own homeownership scatter plot - lable - 'Own'

For mortgage homeownership scatter plot - lable - 'Mortgage'

Label the x-axis as "Annual Income"

and the y-axis as "Loan Amount"
