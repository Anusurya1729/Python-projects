***`Task 9: Find the data of the salesperson who deals with Laptops and sort the result on the basis of Salesperson name in alphabetical order.`***

```python
rwsd.query('Product=="Laptops"').sort_values('Salesperson',ascending=True)
```

***`Task 10: Find the data of the salesperson who deals with Headphones and sort the result on the basis of Region in alphabetical order.`***

```python
rwsd.query('Product=="Headphones"').sort_values('Region',ascending=True)
```

***`Task 11: Find the data of the product with QuantitySold more than or equal to 200 and then sort the data on the Sales value in descending order.`***

```python
rwsd.query('QuantitySold>=200').sort_values('Sales',ascending=False)
```

***`Task 12: Find the data of the product with QuantitySold more than or equal to 200 and then sort the Salesperson name in descending alphabetical  order. Show only Sales value of sorted data.`***

```python
 rwsd.query('QuantitySold>=200').sort_values('Salesperson',ascending=False)["Sales"]
```

***`Task 13: Find the name of the Salesperson whose Region is North and then sort the data according to the product name in alphabetical order. Get only the Salesperson name for the same.`***

```python
rwsd.query('Region=="North"').sort_values('Product',ascending=True)["Salesperson"]
```

***`Task 14: Find the name of the salesperson whose QuantitySold is more than the average QuantitySold and then sort the data according to the region in descending alphabetical order. Get only the name for the same.`***

```python
rwsd.query('QuantitySold>QuantitySold.mean()').sort_values('Region',ascending=False)["Salesperson"]
```

***`Task 15: Find the data where Sales is less than the average Sales of all the salesperson and then sort the data according to the salesperson name in alphabetical order. Get only the QuantitySold for the same.`***

```python
rwsd.query('Sales<Sales.mean()').sort_values('Salesperson',ascending=True)["QuantitySold"]
```
