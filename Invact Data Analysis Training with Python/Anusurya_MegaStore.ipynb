Task 1: Import Pandas Library and create a dataframe named "salesdata" from a given URL and then print it. URL - https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/SalesLocationData.csv

```python
import pandas as pd
salesdata=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/SalesLocationData.csv")
salesdata
```

**Task 11:**

Store the user input for the category in the variable named as "Input_Category". Text to be shown to the user: "Enter the Category Name : "

Filter the sales dataframe for

**Task 12:**

Store the user input for the state in the variable named as "Input_State". Text to be shown to the user "Enter the State Name : "

Filter the Sales dataframe for

Category- "Electronics" and "Clothing"

and Amount greater than 1000

and State to be taken from user input. Store the resultant data in "Sales2" dataframe.

Sort the "Sales2" dataframe by arranging

**Task 13:**

Store the user input for the State in the variable named as "Input_State2". Text to be shown to the user "Enter the State Name : "

Store the user input for the PaymentMode in the variable named as "Input_PaymentMode". Text to be shown to the user "Enter the PaymentMode: "

Filter the sales dataframe for State to be taken from user input and PaymentMode to be taken from user input. Store the resultant data in "Sales3" dataframe.

Sort the "Sales3" by arranging

"Profit" column in descending order

At the end the user input should be-

State (Input_State2) - Enter the name of State: Karnataka

PaymentMode (Input_PaymentMode) - Enter the name of PaymentMode: "Debit Card"

```python
Input_State2=input("Enter the State Name : ")
Input_PaymentMode=input("Enter the PaymentMode: ")
Sales3=salesdata.query('State==@Input_State2 and PaymentMode==@Input_PaymentMode')
Sales3.sort_values('Profit',ascending=False)
```

**Task 14:**

Store the user input for the Category in the variable named as "Input_Category1". Text to be shown to the user "Enter the Category Name : "

Filter the sales dataframe

for City - Bangalore and Chennai

and category to be taken from user input. Store the resultant data in "Sales4" dataframe.

Create a pivot table using the Sales4 dataframe named pivot_2 to show the Sub-Category-wise total Amount and total Profit for each of the category. Use index as sub-category, columns as City and values as Amount and Profit. Print this pivot table. This will show the data for the category selected by the user.

At the end the user input should be-

Category (Input_Category1) - Enter the Category  name : "Clothing"


```python
Input_Category1=input("Enter the Category Name : ")
Sales4=salesdata.query('City in ["Bangalore","Chennai"] and Category==@Input_Category1')
pivot_2=pd.pivot_table(Sales4,index=["Sub-Category"],columns=["City"],values=["Amount","Profit"],aggfunc={"Amount":['sum'],"Profit":['sum']})
pivot_2
```
