**Task 11:**

**Task 13:**

Store the user input for the city in the variable named as "Input_City1". Text to be shown to the user "Enter the City Name : "

Store the user input for the product in the variable named as "Input_Product". Text to be shown to the user "Enter the Product Name : "

Filter the sales dataframe for
City to be taken from user input
and Product to be taken from user input. Store the resultant data in "Sales3" dataframe.

Sort the "Sales3" by arranging

"Sales‚Äù column in descending order

At the end the user input should be-

City (Input_City1) - Enter the name of city: Pune

Product (Input_Product) - Enter the name of product: Sofa

**Task 14:**

Store the user input for the city in the variable named as "Input_City3". Text to be shown to the user "Enter the City Name : "

Store the user input for the Salesperson in the variable named as "Input_Salesperson1". Text to be shown to the user "Enter the Salesperson Name : "
Filter the sales dataframe for
City to be taken from user input
and Salesperson to be taken from user input. Store the resultant data in "Sales4" dataframe.


Create a pivot table using the Sales4 dataframe named pivot_2 to show the Product-wise total sales and total unit for the Salesperson. Use index as Product, columns as Salesperson and values as Sales and Unit. Print this pivot table. This will show the data for the salesperson and city selected by the user.

At the end the user input should be-

City (Input_City3) - Enter the city name : Chennai

Salesperson (Input_Salesperson1) - Enter the Salesperson name : Vipin Sharma

```python
Input_City3=input("Enter the City Name : ")
Input_Salesperson1=input("Enter the Salesperson Name : " )
Sales4=sales.query('City==@Input_City3 and Salesperson==@Input_Salesperson1')
pivot_2=pd.pivot_table(Sales4,index=["Product"],columns=["Salesperson"],values=["Sales","Unit"],aggfunc={"Sales":["sum"],"Unit":["sum"]})
pivot_2
```
