**Task 1**

Import Pandas and matplotlib Library. Create a dataframe named "ftdata" from a given URL and then print it. URL - https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/HW1exDOE11csv.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
ftdata=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/HW1exDOE11csv.csv")
ftdata
```

**Task 2**

In the data the number of impressions, views, comments and likes are provided in numbers which we would like to convert into lakhs. For this you have to convert impressionCount, viewCount, commentCount, and likeCount into lakhs and round to 2 decimal places and name these columns -

impressionCount(in lakhs),  viewCount(in lakhs),  commentCount(in lakhs),  likeCount(in lakhs),  Convert durationSeconds to durationMinutes and round to 2 decimal places.

Display the 5 rows of DataFrame

```python
ftdata["impressionCount(in lakhs)"]=(ftdata["impressionCount"]/100000).round(2)
ftdata["viewCount(in lakhs)"]=(ftdata["viewCount"]/100000).round(2)
ftdata["commentCount(in lakhs)"]=(ftdata["commentCount"]/100000).round(2)
ftdata["likeCount(in lakhs)"]=(ftdata["likeCount"]/100000).round(2)
ftdata["durationMinutes"]=(ftdata["durationSeconds"]/60).round(2)
ftdata.head(5)
```

**Task 3:**


Extract the year and month from publishTime in separate column named "year" and "month" respectively.

Display the 5 rows of DataFrame



```python
ftdata["publishTime"]=pd.to_datetime(ftdata["publishTime"])
ftdata["year"]=ftdata["publishTime"].dt.year
ftdata["month"]=ftdata["publishTime"].dt.month
ftdata.head(5)
```

**Task 4:**

Drop the following columns - 'impressionCount',  'viewCount', 'publishTime', 'commentCount', 'likeCount',  'durationSeconds' as we have calculated new columns for these columns.

Display the 5 rows of DataFrame

```python
ftdata.drop(columns=["impressionCount",	"viewCount",	"publishTime",	"commentCount",	"likeCount","durationSeconds"],inplace=True)
ftdata.head(5)
```

**Task 5:**

Create a pivot table from the ftdata DataFrame that counts the number of videos (title) for each year. Store this pivot table in ftDatapivotbar and reset its index. Then, print the ftDatapivotbar.

```python
ftDatapivotbar=pd.pivot_table(ftdata,index=["year"],values=["title"],aggfunc={"title":"count"})
ftDatapivotbar.reset_index(inplace=True)
ftDatapivotbar
```

**Task 6:**

Create a bar plot using the ftDatapivotbar DataFrame, with years on the x-axis and the count of videos on the y-axis. Then, display the plot.

```python
plt.bar(x=ftDatapivotbar['year'],height=ftDatapivotbar['title'])
plt.show()
```

**Task 7:**

Create a bar plot using the ytDatapivotbar DataFrame, with years on the x-axis and the count of videos on the y-axis.

Label the x-axis as "Year"

and the y-axis as "Count of videos".

Add the title "Yearly wise videos uploaded",

# Scattered

**Task 17:**

Create a scatter plot to visualize the relationship between likeCount and impressionCount for videos categorized as "Long" in the ftdata DataFrame. Then, display the plot.

```python
Long_likeCount=ftdata.query('videoType=="Long"')["likeCount(in lakhs)"]
Long_impressionCount=ftdata.query('videoType=="Long"')["impressionCount(in lakhs)"]
plt.scatter(Long_likeCount,Long_impressionCount)
plt.show()
```

**Task 18:**

Create a scatter plot to visualize the relationship between likeCount and impressionCount for videos categorized as "Shorts" and "Long" in the ftdata DataFrame. Then, display the plot.

```python
Long_likeCount=ftdata.query('videoType=="Long"')["likeCount(in lakhs)"]
Long_impressionCount=ftdata.query('videoType=="Long"')["impressionCount(in lakhs)"]
plt.scatter(Long_likeCount,Long_impressionCount)
Shorts_likeCount=ftdata.query('videoType=="Shorts"')["likeCount(in lakhs)"]
Shorts_impressionCount=ftdata.query('videoType=="Shorts"')["impressionCount(in lakhs)"]
plt.scatter(Shorts_likeCount,Shorts_impressionCount)
plt.show()
```

**Task 19:**

Add the label and titles in the scatter plot created above as per the details provided below:

For shorts scatter plot - legend - 'Shorts'

For Long scatter plot - legend - 'Long'

Label the x-axis as "Like Count(in lakhs)"

and the y-axis as "Impression Count(in lakhs)"

and place the legend on upper left and legend title is 'Video Type'

```python
Long_likeCount=ftdata.query('videoType=="Long"')["likeCount(in lakhs)"]
Long_impressionCount=ftdata.query('videoType=="Long"')["impressionCount(in lakhs)"]
plt.scatter(Long_likeCount,Long_impressionCount,label="Long")
Shorts_likeCount=ftdata.query('videoType=="Shorts"')["likeCount(in lakhs)"]
Shorts_impressionCount=ftdata.query('videoType=="Shorts"')["impressionCount(in lakhs)"]
plt.scatter(Shorts_likeCount,Shorts_impressionCount,label="Shorts")
plt.xlabel("Like Count(in lakhs)")
plt.ylabel("Impression Count(in lakhs)")
plt.title("Shorts vs Long")
plt.legend(title="legend",loc='upper left')
plt.show()
```
