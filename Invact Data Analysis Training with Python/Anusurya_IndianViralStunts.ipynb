You have been given data of a YouTube channel "IndianViralStunts". As a data analyst, you're tasked with uncovering the secrets of viral YouTube videos for the channel. Using the given dataset, you'll clean and visualize the data to reveal trends and correlations. By plotting impressions, views, comments, and likes against video duration and type, you'll understand what drives user engagement. Employ plt.plot, plt.scatter, plt.pie, and plt.bar to create compelling visuals that highlight these insights.

**Task 1**


Import Pandas  and matplotlib Library.
Create a dataframe named "ytdata" from a given URL and then print it.

URL - https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/ytdata%20-%20Sheet1.csv

```python
import pandas as pd
import matplotlib.pyplot as plt
ytdata=pd.read_csv(" https://raw.githubusercontent.com/Invact-Abhay/DOE11/main/ytdata%20-%20Sheet1.csv")
ytdata
```

**Task 2**

In the data the number of impressions, views, comments and likes are provided in numbers which we would like to convert into lakhs. For this you have to convert impressionCount, viewCount, commentCount, and likeCount into lakhs and round to 2 decimal places  and name these columns  -


*   impressionCount(in lakhs)
*   viewCount(in lakhs)
*   commentCount(in lakhs)
*   likeCount(in lakhs)


Convert durationSeconds to durationMinutes and round to 2 decimal places.

Display the 5 rows of DataFrame

```python
ytdata['impressionCount(in lakhs)']=(ytdata['impressionCount']/100000).round(2)
ytdata['viewCount(in lakhs)']=(ytdata['viewCount']/100000).round(2)
ytdata['commentCount(in lakhs)']=(ytdata['commentCount']/100000).round(2)
ytdata['likeCount(in lakhs)']=(ytdata['likeCount']/100000).round(2)
ytdata['durationMinutes']=(ytdata['durationSeconds']/60).round(2)
ytdata.head(5)
```

**Task 3:**

Extract the year and month from publishTime in separate column named "year" and "month" respectively.

Display the 5 rows of DataFrame

```python
ytdata['publishTime']=pd.to_datetime(ytdata['publishTime'])
ytdata['month']=ytdata['publishTime'].dt.month
ytdata['year']=ytdata['publishTime'].dt.year
ytdata.head(5)
```

**Task 4:**


Drop the following columns -
'impressionCount', 'viewCount','publishTime', 'commentCount', 'likeCount', 'durationSeconds' as we have calculated new columns for these columns.

Display the 5 rows of DataFrame

```python
ytdata.drop(columns=['impressionCount', 'viewCount','publishTime', 'commentCount', 'likeCount', 'durationSeconds'],inplace=True)
ytdata.head(5)
```

# Bar Graph


**Task 5:**


Create a pivot table from the ytdata DataFrame that counts the number of videos (id) for each year. Store this pivot table in pivot_ytdata_bar and reset its index.

```python
pivot_ytdata_bar=pd.pivot_table(ytdata,index=['year'],values=['id'],aggfunc={'id':'count'})
pivot_ytdata_bar.reset_index(inplace=True)
pivot_ytdata_bar
```


**Task 6:**

Create a bar graph using the pivot_ytdata_bar DataFrame, with years on the x-axis and the count of videos on the y-axis. Then, display the graph.

```python
plt.bar(x=pivot_ytdata_bar['year'],height=pivot_ytdata_bar['id'])
plt.show()
```

**Task 7:**

Create a bar graph using the pivot_ytdata_bar DataFrame, with years on the x-axis and the count of videos on the y-axis.

Label the x-axis as "Year"

and the y-axis as "Count of videos".

Add the title "Yearly wise videos uploaded",

and rotate the x-axis labels by 45 degrees.

Then, display the graph.

```python
plt.bar(x=pivot_ytdata_bar['year'],height=pivot_ytdata_bar['id'])
plt.xlabel('year')
plt.ylabel('Count of videos')
plt.title("Yearly wise videos uploaded")
plt.xticks(ticks=pivot_ytdata_bar['year'],rotation=45)
plt.show()
```

# Scatter Plot


**Task 17:**

Create a scatter plot to visualize the relationship between view count and impression count for videos categorized as "Shorts" in the ytdata DataFrame. Then, display the plot.

```python
view_count=ytdata.query('videoType=="Shorts"')['viewCount(in lakhs)']
impression_count=ytdata.query('videoType=="Shorts"')['impressionCount(in lakhs)']
plt.scatter(view_count,impression_count)
plt.show()
```

**Task 18:**

Create a scatter plot to visualize the relationship between view count and impression count for videos categorized as "Shorts" and "Long" in the ytdata DataFrame. Then, display the plot.








```python
view_count=ytdata.query('videoType=="Shorts"')['viewCount(in lakhs)']
impression_count=ytdata.query('videoType=="Shorts"')['impressionCount(in lakhs)']
plt.scatter(view_count,impression_count)
view_count=ytdata.query('videoType=="Long"')['viewCount(in lakhs)']
impression_count=ytdata.query('videoType=="Long"')['impressionCount(in lakhs)']
plt.scatter(view_count,impression_count)
plt.show()
```

**Task 19:**


Add the label and titles in the scatter plot created above as per the details provided below:

For shorts scatter plot -   legend - 'Shorts'

For Long scatter plot -   legend - 'Long'

Label the x-axis as "Impression Count(in lakhs)"

and the y-axis as "View Count(in lakhs)"

Add the title "Shorts vs Long".

and place the legend on upper left and legend title is 'Video Type'

Then, display the plot.


```python
view_count=ytdata.query('videoType=="Shorts"')['viewCount(in lakhs)']
impression_count=ytdata.query('videoType=="Shorts"')['impressionCount(in lakhs)']
plt.scatter(view_count,impression_count,label='Shorts')
view_count=ytdata.query('videoType=="Long"')['viewCount(in lakhs)']
impression_count=ytdata.query('videoType=="Long"')['impressionCount(in lakhs)']
plt.scatter(view_count,impression_count,label='Long')
plt.xlabel("Impression Count(in lakhs)")
plt.ylabel("View Count(in lakhs)")
plt.title("Shorts vs Long")
plt.legend(title='Video Type',loc='upper left')
plt.show()
```
