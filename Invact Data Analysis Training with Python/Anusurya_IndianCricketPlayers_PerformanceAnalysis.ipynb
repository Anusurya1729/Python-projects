***`Task 9: Find the data of the teams who have played less than 100 matches and sort the result on the basis of Team Name in alphabetical order.`***

```python
splr.query('MatchesPlayed<100').sort_values('Team',ascending=True)
```

***`Task 10: Find the data of the teams who have BattingAverage of 58.92 and sort the result on the basis of Team's name in the descending alphabetical order.`***

```python
splr.query('BattingAverage==58.92').sort_values('Team',ascending=False)
```

***`Task 11: Find the data of the teams with HighestScore more than 260 and then sort the data on the Matchesplayed in descending order.`***

```python
splr.query('	HighestScore>260').sort_values('MatchesPlayed',ascending=False)
```

***`Task 12: Find the data of the teams that have RunsScored  more than 10000 and then sort the data on the HighestScore in descending order. Get only the Player names for the same.`***

```python
splr.query('RunsScored>10000').sort_values('HighestScore',ascending=False)["PlayerName"]
```

***`Task 13: Find the name of the teams whose RunsScored is less than the Average RunsScored of all the teams and then sort the data according to the Player’s name in the alphabetical order. `***

```python
splr.query('RunsScored<RunsScored.mean()').sort_values('PlayerName',ascending=True)
```
