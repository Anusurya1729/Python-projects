***`Task 1: Import the pandas library `***

```python
import pandas as pd
```

***`Task 2: Create a dataframe named “fplr” from a given URL.`***

https://raw.githubusercontent.com/Invact-Abhay/DOE/main/DOE2HW3.csv

```python
fplr=pd.read_csv("https://raw.githubusercontent.com/Invact-Abhay/DOE/main/DOE2HW3.csv")
fplr
```

***`Task 3: Sort the data according to the Position in the alphabetical order.`***

```python
fplr.sort_values('Position',ascending=True)
```

***`Task 4: Sort the data according to the Team's name in the descending alphabetical order.`***

```python
fplr.sort_values('Team',ascending=False)
```

***`Task 5: Sort the data according to the GoalsScored in the ascending order.`***

```python
fplr.sort_values('GoalsScored',ascending=True)
```

***`Task 6: Sort the data according to the ‘Assists’ in the descending order.`***

```python
fplr.sort_values('Assists',ascending=False)
```

***`Task 7: Sort the data according to the Player's name in the descending alphabetical order and only show the Player's name.`***

```python
fplr.sort_values('PlayerName',ascending=False)["PlayerName"]
```

***`Task 8: Sort the data based on the TotalShots in ascending order and give only the name of the teams of the sorted data.`***

```python
fplr.sort_values('TotalShots',ascending=True)["Team"]
```

***`Task 9: Find the data of the teams who have PassAccuracyPercent more than 90 and sort the result on the basis of Team Name in alphabetical order.`***

```python
fplr.query('PassAccuracyPercent>90').sort_values('Team',ascending=True)
```

***`Task 10: Find the data of the teams who have Tackles more than 50 and sort the result on the basis of Player’s name in the descending alphabetical order.`***

```python
fplr.query('Tackles>50').sort_values('PlayerName',ascending=False)
```

***`Task 11: Find the data of the teams with Assists less than 5 and then sort the data on the GoalsScored in descending order.`***

```python
fplr.query('Assists<5').sort_values('GoalsScored',ascending=False)
```

***`Task 12: Find the data of the teams that have GoalsScored more than 15 and then sort the data on the Tackles in ascending order. Get only the Player names for the same.`***

```python
fplr.query('GoalsScored>15').sort_values('Tackles',ascending=True)["PlayerName"]
```

***`Task 13: Find the name of the teams whose GoalsScored is less than the Average GoalsScored of all the teams and then sort the data according to the Player’s name in the alphabetical order.`***

```python
fplr.query('GoalsScored<GoalsScored.mean()').sort_values('PlayerName',ascending=True)
```

***`Task 14: Find the name of the players whose Assists is less than the average Assists and then sort the data according to the Players name in descending alphabetical order. Get only the player’s name for the same.`***

```python
fplr.query('Assists<Assists.mean()').sort_values('PlayerName',ascending=False)["PlayerName"]
```

***`Task 15: Find the name of the players whose TotalShots is more than the average TotalShots of all the players and then sort the data according to the Player’s name in alphabetical order. Get only the player names for the same.`***

```python
fplr.query('TotalShots>TotalShots.mean()').sort_values('PlayerName',ascending=True)["PlayerName"]
```
