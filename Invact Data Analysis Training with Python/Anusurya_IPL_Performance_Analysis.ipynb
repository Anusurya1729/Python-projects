***`Task 2: Create a dataframe named “ipl” from a given URL.`***

https://raw.githubusercontent.com/Anpr1211/DOE_Datasets/main/DOE3_IPL.csv

```python
ipl=pd.read_csv("https://raw.githubusercontent.com/Anpr1211/DOE_Datasets/main/DOE3_IPL.csv")
ipl
```

***`Task 3: Sort the data according to the Zone in the alphabetical order.`***

```python
ipl.sort_values('Zone',ascending=True)
```

***`Task 4: Sort the data according to the Team's name in the descending alphabetical order.`***

```python
ipl.sort_values('Teams',ascending=False)
```

***`Task 5: Sort the data according to the Total Matches played in the ascending order.`***

```python
ipl.sort_values('TotalMatches',ascending=True)
```

***`Task 6: Sort the data according to the Total Matches played in the descending order.`***

```python
ipl.sort_values('TotalMatches',ascending=False)
```

***`Task 7: Sort the data according to the Team's name in the descending alphabetical order and only show Team names.`***

```python
ipl.sort_values('Teams',ascending=False)["Teams"]
```

***`Task 8: Sort the data based on the Win Percent in ascending order and give only the name of the teams of the sorted data.`***

```python
ipl.sort_values('WinPercent',ascending=True)["Teams"]
```

***`Task 9: Find the data of the teams who have played more than or equal to 100 matches and sort the result on the basis of Team Name in alphabetical order.`***

```python
ipl.query('TotalMatches>=100').sort_values('Teams',ascending=True)
```

***`Task 10: Find the data of the teams who have lost exactly 50 matches and sort the result on the basis of Team Name in alphabetical order.`***

```python
ipl.query('MatchesLost==50').sort_values('Teams',ascending=True)
```

***`Task 11: Find the data of the teams that have a win percent of more than 50 and then sort the data on the total matches played in descending order.`***

```python
ipl.query('WinPercent>50').sort_values('TotalMatches',ascending=False)
```

***`Task 12: Find the data of the teams that have a win percent of more than 40 and then sort the data on the total matches played in descending order. Get only the team names for the same.`***

```python
ipl.query('WinPercent>40').sort_values('TotalMatches',ascending=False)["Teams"]
```

***`Task 13: Find the name of the teams whose WinPercent is less than the average WinPercent of all the teams and then sort the data according to the Team's name in the alphabetical order. `***

```python
ipl.query('WinPercent<WinPercent.mean()').sort_values('Teams',ascending=True)
```

***`Task 14: Find the name of the teams whose total matches is less than the average total matches and then sort the data according to the Team's name in alphabetical order. Get only the team names for the same.`***

```python
ipl.query('TotalMatches<TotalMatches.mean()').sort_values('Teams',ascending=True)["Teams"]
```
